## Arquitectura Empresarial

Un sistema coherente de principios, métodos y modelos que se usa para diseñar y construir la estructura de una organización, sus procesos de negocio, sus sistemas de información y su infraestructura […] Captura los elementos esenciales del negocio, de TI y de su evolución, y proporciona una visión holística.

![](https://github.com/jazaineam1/BigData2023_2/blob/main/Images/2.1.png?raw=true)



La arquitectura empresarial generalmente abarca varios dominios clave:

1. **Arquitectura Empresarial de Negocios (Business Architecture)**: Define cómo opera la organización, incluyendo su misión, visión, estrategias, objetivos, procesos y modelos de negocio. Ayuda a asegurar que las actividades de la empresa estén alineadas con su estrategia y objetivos.

2. **Arquitectura Empresarial de Tecnología (Technology Architecture)**: Se refiere a la infraestructura tecnológica, sistemas y aplicaciones que soportan las operaciones de la empresa. Esto incluye la selección de tecnologías, estándares, plataformas y la integración de sistemas.

3. **Arquitectura Empresarial de Información (Data Architecture)**: Trata con la gestión y organización de los datos dentro de la empresa. Esto incluye la definición de estructuras de datos, flujos de datos, almacenamiento y procesamiento de datos.

4. **Arquitectura Empresarial de Aplicaciones (Application Architecture)**: Se enfoca en cómo se desarrollan, implementan y mantienen las aplicaciones de software en la organización. Esto implica decisiones sobre las tecnologías de desarrollo, la integración de sistemas y la gestión del ciclo de vida de las aplicaciones.

## Business Process Management (BPM)

![](https://decidesoluciones.es/wp-content/uploads/2020/05/BPMLC_reduced_1-3.png)

El Business Process Management (BPM), o Gestión de Procesos de Negocio en español, es un enfoque y conjunto de prácticas destinado a mejorar y optimizar los procesos operativos de una organización. Consiste en identificar, diseñar, automatizar, ejecutar, medir, controlar y mejorar continuamente los procesos empresariales con el objetivo de aumentar la eficiencia, la calidad, la agilidad y la capacidad de respuesta de la empresa.

BPM implica trabajar en la comprensión y el análisis de los procesos existentes, identificar oportunidades de mejora, implementar cambios y medir el impacto de esos cambios en términos de eficiencia, calidad y satisfacción del cliente. La automatización también es un componente clave del BPM, donde se utilizan herramientas tecnológicas para agilizar la ejecución de tareas y el flujo de información.

La relación entre Business Process Management (BPM) y la arquitectura del negocio es estrecha, ya que ambas disciplinas están relacionadas con la gestión y optimización de cómo opera una organización en su conjunto.

Aquí hay un ejemplo de BPM en acción:

**Ejemplo: Proceso de Gestión de Solicitudes de Reembolso en una Compañía de Seguros**

Imagina una compañía de seguros que recibe solicitudes de reembolso de los clientes. Antes de implementar BPM, el proceso puede ser manual, desorganizado y propenso a errores. Los empleados pueden perder tiempo buscando información, los tiempos de respuesta pueden variar y la calidad de servicio podría ser inconsistente.

Con BPM, el proceso se puede mejorar de la siguiente manera:

1. **Diseño y Documentación**: Se analiza el proceso actual de gestión de reembolsos. Se identifican los pasos involucrados, las personas responsables y los puntos de contacto con los clientes.

2. **Optimización**: Se identifican cuellos de botella, duplicaciones y demoras en el proceso. Se diseñan soluciones para abordar estas ineficiencias.

3. **Automatización**: Se implementa un sistema de software BPM que automatiza la creación y el seguimiento de las solicitudes de reembolso. Los clientes pueden enviar solicitudes en línea, que son dirigidas automáticamente a los departamentos correspondientes.

4. **Rastreo y Monitoreo**: Se pueden rastrear las solicitudes en tiempo real. Los empleados pueden ver el estado de cada solicitud y saber en qué etapa se encuentra.

5. **Aprobaciones y Notificaciones**: El sistema BPM puede enviar notificaciones automáticas a los empleados responsables de revisar y aprobar las solicitudes. Esto agiliza el proceso y reduce la posibilidad de retrasos.

6. **Medición y Mejora**: Se recopilan datos sobre el tiempo que lleva procesar cada solicitud, la calidad de las respuestas y la satisfacción del cliente. Estos datos se utilizan para identificar oportunidades adicionales de mejora.

Mediante el uso de BPM, la compañía de seguros logra una mayor eficiencia en el proceso de reembolsos. Los tiempos de respuesta son más cortos, la información es más precisa y la experiencia del cliente mejora notablemente. Además, el enfoque en la mejora continua permite seguir ajustando el proceso para adaptarse a las necesidades cambiantes de la empresa y del mercado.

## Ciclo de vida de los datos

![](https://scontent.fbog3-1.fna.fbcdn.net/v/t1.6435-9/120273370_652092392159862_5896927628537707756_n.jpg?stp=dst-jpg_p526x296&_nc_cat=100&ccb=1-7&_nc_sid=8bfeb9&_nc_ohc=rLs0C7FE78oAX-dtJR1&_nc_ht=scontent.fbog3-1.fna&oh=00_AfA4vA8RldZ-BOf5iPHm2PQezjycuR4m0aIKoF_XinYudQ&oe=65186971)

¿Qué es el ciclo de vida de los datos?

El ciclo de vida de los datos es una secuencia de etapas por las que pasan los datos a lo largo de toda su vida útil.

Los datos se separan en fases en función de diferentes criterios, y pasan por estas etapas a medida que completan diferentes tareas o cumplen ciertos requisitos.

El ciclo de vida de los datos abarca todo el periodo de tiempo que los datos existen en una organización, desde la generación de los datos hasta su eliminación o reutilización a través de diferentes tipos de repositorios de investigación.

Se considera que es un ciclo porque los conocimientos obtenidos de un proyecto de datos suelen servir de base para el siguiente. De este modo, la última etapa del proceso retroalimenta la primera.

Tener una visión clara de las etapas en las que se encuentran los datos permite a las organizaciones manejarlos con mayor eficiencia y seguridad.

En el proceso de análisis nos podemos encontrar con diferentes herramientas que pueden ser de gran ayuda para la generación de conocimiento.

**Minería de Datos:**

<img src="https://cdn-icons-png.flaticon.com/512/1998/1998537.png" alt="Mi imagen" width="230">

La minería de datos, también conocida como descubrimiento de datos, es una forma especializada de análisis de datos que se dirige a grandes conjuntos de datos.

En relación con el análisis de Big Data, la minería de datos generalmente se refiere a técnicas automatizadas basadas en software que tamizan conjuntos de datos masivos para identificar patrones y tendencias.

La minería de datos constituye la base para analítica predictiva e inteligencia empresarial (BI).

**Análisis Estadístico:**

<img src="https://cdn-icons-png.flaticon.com/512/786/786695.png" alt="Mi imagen" width="230">

 El análisis estadístico utiliza métodos estadísticos basados ​​en fórmulas matemáticas como medio para analizar datos.Este tipo de análisis se usa comúnmente para describir conjuntos de datos a través de resúmenes, como proporcionando la media, la mediana o la moda de las estadísticas asociadas con el conjunto de datos.

También puede utilizarse para inferir patrones y relaciones dentro del conjunto de datos, como regresión y correlación.

* Pruebas A/B: Las pruebas A/B, también conocidas como pruebas divididas o de cubo, comparan dos versiones de un elemento para determinar qué versión es superior en función de una métrica predefinida.

>Por ejemplo, para determinar el mejor diseño posible para un anuncio de helado en el sitio web de la empresa A, se utilizan dos versiones diferentes del anuncio. La versión A es una existente ad (el control) mientras que la Versión B ha tenido su diseño ligeramente alterado (el tratamiento). Luego, las versiones se muestran simultáneamente a diferentes usuarios:

 >Versión A al Grupo A

 >Versión B al Grupo B

 * Correlación: La correlación es una técnica de análisis utilizada para determinar si dos variables están relacionadas entre sí.

>La correlación es por lo tanto comúnmente utilizado para la minería de datos donde la identificación de las relaciones entre variables en un conjunto de datos conduce al descubrimiento de patrones y anomalías.

>Por ejemplo, los gerentes creen que las heladerías necesitan almacenar más helados para los clientes en días calientes, pero no sé cuánto más almacenar.

* Regresión: La técnica de análisis de regresión explora cómo una variable dependiente se relaciona con una variable independiente dentro de un conjunto de datos.

>La aplicación de esta técnica ayuda a determinar cómo cambia el valor de la variable dependiente en relación con los cambios en el valor de la variable independiente.

>Por ejemplo, para determinar cuánto stock adicional necesita cada heladería, los analistas aplican la regresión ingresando los valores de las lecturas de temperatura. Estos se basan en el pronóstico del tiempo como variable independiente y la cantidad de cremas vendidas como variable dependiente. Lo que descubren los analistas es que el 15% de se requiere stock adicional por cada aumento de temperatura de 5 grados.

 Los humanos son buenos para detectar patrones y relaciones dentro de los datos. Desafortunadamente no pueden procesar grandes cantidades de datos muy rápidamente.

Tipos de técnicas de aprendizaje automático:
Clasificación
Agrupación
Detección de valores atípicos
Filtrado

* Calsificación Supervisado:  La clasificación es una técnica de aprendizaje supervisado mediante la cual los datos se clasifican en relevantes, categorías previamente aprendidas. Consiste en dos pasos:

>El sistema recibe datos de entrenamiento que ya están categorizados o etiquetados, para que pueda desarrollar una comprensión de las diferentes categorías.

>El sistema recibe datos desconocidos pero similares para la clasificación y se basa en la entendiéndolo desarrollado a partir de los datos de entrenamiento, el algoritmo clasificará los datos sin etiquetar.

* Calsificación No  Supervisado: El agrupamiento es una técnica de aprendizaje no supervisado mediante la cual los datos se dividen en diferentes grupos para que los datos de cada grupo tengan propiedades similares.

>No hay un aprendizaje previo de categorías requeridas.

>La agrupación en clústeres se usa generalmente en la minería de datos para obtener una comprensión de las propiedades de un conjunto de datos dado.

>Después de desarrollar esta comprensión, la clasificación se puede utilizar para hacermejores predicciones sobre datos similares pero nuevos o no vistos.
